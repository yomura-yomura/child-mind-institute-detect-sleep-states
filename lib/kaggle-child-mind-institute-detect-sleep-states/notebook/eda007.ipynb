```python
import numpy as np
from pathlib import Path
import  matplotlib.pyplot as plt
import child_mind_institute_detect_sleep_states.data.comp_dataset as c_data
path_data =Path("/home/jumpei.uchida/develop/child-mind-institute-detect-sleep-states/lib/kaggle-child-mind-institute-detect-sleep-states/output/prepare_data/train/robust_scaler/")
df_event = c_data.get_event_df("train")
```

```python
s_id = "0a96f4993bd7"
df_true = df_event.query("series_id == @s_id")
df_onset = df_true.query("event == 'onset'")
df_wakeup = df_true.query("event == 'wakeup'")
```

```python
file = "n_unique_anglez_5min.npy"
path_file = path_data / s_id / file
arr = np.load(path_file)
plt.figure(figsize =(15,5))
plt.plot(arr,color = "blue",label="feature")
plt.vlines(df_wakeup["step"].values, 0, 1, label="actual_wakeup", linestyles="dashed", color="C2",linewidth=4)
plt.vlines(df_onset["step"].values, 0, -1, label="actual_onset", linestyles="dashed", color="C1",linewidth=4)
plt.legend()
plt.show()
```
