```python
import numpy as np
from pathlib import Path
import  matplotlib.pyplot as plt
path_data =Path("/home/jumpei.uchida/develop/child-mind-institute-detect-sleep-states/data/pseudo_label")
```

```python

list_path = []
count = 0
th = 5
for p in path_data.iterdir():
    list_path.append(p)
    data = np.load(p)
    label_data  = data["label"]

    pseudo_data = data["pseudo"]
    if np.all(sum(label_data) == sum(pseudo_data)):
        continue

    count += 1
    sleep_label = label_data[:,0]
    sleep_pseudo = pseudo_data[:,0]
    onset_label = label_data[:,1]
    onset_pseudo = pseudo_data[:,1]
    wakeup_label = label_data[:,2]
    wakeup_pseudo = pseudo_data[:,2]
    if count == th:
        break
```

```python
plt.plot(sleep_pseudo,label = "pseudo",color = "blue")
plt.plot(sleep_label,label = "true",color = "red")
plt.legend()
```

```python

plt.plot(onset_pseudo,color = "blue",label = "pseudo")
plt.plot(onset_label,color = "red",label = "true")
plt.legend()
```

```python

plt.plot(wakeup_pseudo,color = "blue",label = "pseudo")
plt.plot(wakeup_label,color = "red",label = "true")
```
